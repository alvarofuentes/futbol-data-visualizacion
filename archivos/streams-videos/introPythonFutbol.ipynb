```python
# Nociones basicas
# Términos estadísticos basicos --> que es un promedio/media, mediana, distribuciones, sesgo, varianza, percentiles.
# Nociones basicas de terminos estadísticos en el fútbol
# Nociones basicas de analisis de datos.
# Minimo interes en programar o manejar sheets/excel.
# Diferencia entre analista de datos y cientifico de datos.
```

```python
df = pd.read_csv('archivos/pasesPremier.csv')
df.columns = df.iloc[0].values
df = df.drop(0)
df.Edad = df.Edad.apply(lambda x: x.split('-')[0])
df.Edad = df.Edad.astype(int)
```

```python
df.Edad.mean()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.histplot(df.Edad, bins=5)
plt.axvline(np.percentile(df.Edad, 50), color='r', linestyle='--', label='Mediana')
plt.axvline(np.percentile(df.Edad, 95), color='green', linestyle='-.', label='Percentil 95')
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
df.xA = df.xA.astype(float)
df.PC = df.PC.astype(float)
plt.scatter(df.xA, df.PC)
plt.axvline(np.percentile(df.xA, 50), color='r', linestyle='--', label='Mediana')
plt.axvline(np.percentile(df.xA, 95), color='green', linestyle='-.', label='Percentil 95')
```
