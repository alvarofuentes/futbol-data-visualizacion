```python
messi = df_partido[df_partido['player'] == 'Lionel Andrés Messi Cuccittini']
messi 
```

```python
messi.columns
```

```python
messi_filt = messi[['location', 'pass_end_location', 'pass_outcome', 'pass_recipient', 'carry_end_location', 'shot_end_location', 'shot_outcome', 'shot_statsbomb_xg']].reset_index(drop=True)
```

```python
messi_filt[['pass_end_x', 'pass_end_y']] = messi_filt['pass_end_location'].apply(pd.Series)
messi_filt[['x', 'y']] = messi_filt['location'].apply(pd.Series)
messi_filt[['carry_end_x', 'carry_end_y']] = messi_filt['carry_end_location'].apply(pd.Series)
messi_filt[['shot_end_x', 'shot_end_y', 'shot_end_z']] = messi_filt['shot_end_location'].apply(pd.Series)
```

```python
fig = plt.figure(figsize = (16,9), dpi = 100)
ax = plt.subplot(111)
pitch = Pitch(
    pitch_type='statsbomb',
    axis=True, 
    label=True, 
    tick=True,
    goal_type='box'
)
pitch.draw(ax = ax)
pitch.scatter(messi_filt['x'], messi_filt['y'],ax=ax)
```

```python
fig = plt.figure(figsize = (16,9), dpi = 100)
ax = plt.subplot(111)
pitch = Pitch(
    pitch_type='statsbomb',
    goal_type='box'
)
pitch.draw(ax = ax)
pases_mess_corr = messi_filt[~messi_filt['pass_outcome'].isna() == True]
pases_mess_err = messi_filt[messi_filt['pass_outcome'].isna() == True]

pitch.arrows(pases_mess_corr['x'], pases_mess_corr['y'], pases_mess_corr['pass_end_x'], pases_mess_corr['pass_end_y'],ax=ax, width=2,
             headwidth=3.5, color='green', label='Pases completos')
pitch.arrows(pases_mess_err['x'], pases_mess_err['y'], pases_mess_err['pass_end_x'], pases_mess_err['pass_end_y'],ax=ax, width=2,
             headwidth=3.5, color='red', label='Pases errados')
plt.legend(loc='lower center', ncol=2)
```

```python
fig = plt.figure(figsize = (16,9), dpi = 100)
ax = plt.subplot(111)
pitch = VerticalPitch(
    pitch_type='statsbomb',
    goal_type='box',
    half=True
)
pitch.draw(ax = ax)
shot_mess_corr = messi_filt[(messi_filt['shot_outcome'] != 'Goal') & (messi_filt['shot_statsbomb_xg'].isna() == False)]
shot_mess_err = messi_filt[messi_filt['shot_outcome'] == 'Goal']

pitch.scatter(shot_mess_corr['x'], shot_mess_corr['y'],s=shot_mess_corr['shot_statsbomb_xg']*350,ax=ax, color ='red', label='No gol')
pitch.scatter(shot_mess_err['x'], shot_mess_err['y'],s=shot_mess_err['shot_statsbomb_xg']*350,ax=ax, marker='football', label='Gol')

plt.legend(loc='lower center', ncol=2)
```
